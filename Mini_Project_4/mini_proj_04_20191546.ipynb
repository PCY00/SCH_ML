### 시드 찾기

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import cross_val_score
import numpy as np

best_seed = None
best_mean_rmse = np.inf

# 100번의 시드를 적용하여 교차 검증 실행
for i in range(0, 2880 * 123, 123):
    seed = 42 + i  # 시드 값 설정
    np.random.seed(seed)
    sh_in = np.random.permutation(len(wine))
    sh_t_data = wine[sh_in]
    sh_t_target = wine_labels.values[sh_in]

    extra_trees_clf = ExtraTreesRegressor(
        max_depth=31, max_features=3, min_samples_leaf=1, min_samples_split=2,
        n_estimators=1580, random_state=42
    )

    scores = cross_val_score(
        extra_trees_clf, sh_t_data, sh_t_target, scoring="neg_mean_squared_error", cv=10, n_jobs=-1
    )
    rmse_scores = np.sqrt(-scores)
    mean_rmse = np.mean(rmse_scores)

    print(f"Seed: {seed}, Mean RMSE: {mean_rmse}")

    if mean_rmse < best_mean_rmse:
        best_mean_rmse = mean_rmse
        best_seed = seed

print(f"Best Seed: {best_seed}, Best Mean RMSE: {best_mean_rmse}")
```
